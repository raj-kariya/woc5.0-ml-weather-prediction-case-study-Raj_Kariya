```python
from google.colab import drive 
drive.mount('/content/gdrive')
import pandas as pd
df2 = pd.read_csv("gdrive/My Drive/Colab Notebooks/Youtube Influencer Analysis - Updated.csv", encoding='cp1252')

```

Counting the missing values in each column

```python
df2.isnull().sum()
```

```python
Y = df2.fillna(value = 0)
```

```python
df2.dropna()
```

Dropping the columns which are irrelevant features in the model.

```python
Arr = ["Creator Gender","Video Description","Community Engagement (Posts per week)","No of Playlist","Subtitle","Date of the Last Comment"]
df2.drop(Arr,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
 
le = LabelEncoder()
Arr=['Video Link','Video Title','Channel URL','Creator Name','Language of the Video','Intern Who Collected the Data','Date of Video Upload','Duration of Video','Premiered or Not']

for i in Arr:
     label = le.fit_transform(df2[i])
     Y.drop(i,axis=1,inplace=True)
     Y[i]=label

Y
```
