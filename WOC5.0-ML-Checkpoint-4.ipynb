```python
print(mat.style.available)
mat.style.use("dark_background")
x = [10,20,30,40,50,60,70,80,90]
x_ind=np.arange(len(x))
wid=0.25

y = [200,300,400,500,600,700,800,900,1000]
mat.bar(x_ind,y,width=wid)

z=[250,350,450,500,450,550,650,700,850]
mat.bar(x_ind+wid,z,width=wid)

mat.legend(['R Score','D Score'])
mat.bar(x_ind+wid,z,width=wid)
mat.xticks(ticks=x_ind,label=x)
mat.title('Score Board')
mat.xlabel('#Balls')
mat.ylabel('#Runs')
mat.tight_layout()
mat.show()
```

```python
car_de = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CarPrice_Assignment.csv")
car_de
```

```python
car_id = car_de['car_ID']
car_price = car_de['price']
car_horsepower = car_de['horsepower']
car_peakrpm = car_de['peakrpm']
size = car_de['enginesize']

col = np.random.randint(10,size=(205))

x_ind=np.arange(len(car_id))
wid=0.25
mat.style.use("dark_background")

mat.scatter(car_id,car_price,s=size,cmap='Greens',edgecolor="white",linewidth=1,alpha=0.75)
color_b=mat.colorbar()
mat.tight_layout()

mat.xlabel("CAR_ID");
mat.ylabel("PRICE")
```

```python
mat.scatter(car_id,car_peakrpm,s=size,c=col,cmap='Wistia',edgecolor="black",linewidth=1,alpha=0.75)

color_b=mat.colorbar()
mat.tight_layout()

mat.xlabel("CAR_ID");
mat.ylabel("Peakrpm")
```
